# Dictionaries
- Key-Value pairs
- Keys must be unique
- Keys have to be "immutable" objects
    - list is an example not allowed

```python
world = {
    "afghanistan": 30.55,
    "albania": 2.77,
    "algeria": 39.21
}
world["albania"]
```

```python
# set
world["sealand"] = 0.000027
world
```

```python
"sealand" in world
```

```python
# update
world["sealand"] = 0.000028
world
```

```python
del(world["sealand"])
world
```
