# Pandas
- Good for handling dataset with different data types

```python
import pandas as pd
```

## DataFrame

### Create

#### Dictionary

```python
dict = { 
    "country": ["Brazil", "Russia", "India", "China", "South Africa"],
    "capital": ["Brasilia", "Moscow", "New Delhi", "Beijing", "Pretoria"],
    "area": [8.516, 17.10, 3.286, 9.597, 1.221],
    "population": [200.4, 143.5, 1252, 1357, 52.98]
}
brics = pd.DataFrame(dict)
brics.head()
```

```python
brics.index = ["BR", "RU", "IN", "CH", "SA"]
brics.head()
```

#### Import

```python
brics = pd.read_csv('./datasets/brics.csv', index_col=0)
brics.head()
```

### Index and select data
- Square brackets
- `loc` and `iloc` methods

#### Square brackets []

```python
brics['country']
```

```python
type(brics['country'])
```

```python
brics[['country']]
```

```python
brics[['country', 'capital']]
```

```python
brics[1:4]
```

#### loc

```python
brics.loc["RU"] # Series
```

```python
brics.loc[['RU']] # DataFrame
```

```python
brics.loc[['RU', 'IN', 'CH']]
```

```python
brics.loc[['RU', 'IN', 'CH'], ["country", "capital"]]
```

```python
brics.loc[:, ["country", "capital"]]
```

`loc` is more versatile than square brackets for row & column access.

#### iloc

```python
brics.iloc[1] # Series
```

```python
brics.iloc[[1]] # DataFrame
```

```python
brics.iloc[[1, 2, 3], [0, 1]]
```

```python
brics.iloc[:, [0, 1]]
```

## Filtering

```python
is_huge = brics["area"] > 8
brics[is_huge]
```

```python
import numpy as np
brics[np.logical_and(brics["area"] > 8, brics["area"] < 10)]
```

## Loops

```python
for lab, row in brics.iterrows():
    print(lab + ": " + row["capital"])
```

```python
for lab, row in brics.iterrows():
    # - Creating Series on every iteration
    brics.loc[lab, "name_length"] = len(row["country"])
print(brics)
```

```python
# more efficient
brics["name_length"] = brics["country"].apply(len)
print(brics)
```
